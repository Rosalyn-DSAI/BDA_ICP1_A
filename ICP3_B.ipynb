```python
# Some matplotlib workflow (Simple line plot)
# Import and get matplotlib ready
import matplotlib.pyplot as plt

# Prepare data (create two lists of 5 numbers, X & y)
X = [2,4,6,8,10]
Y = [10,20,30,40,50]

# Setup figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Add data (X, y) to axes
ax.plot(X,Y)

# Customize plot by adding a title, xlabel and ylabel
ax.set(title='Simple Plot', xlabel='x', ylabel='y')

# Save the plot to file using fig.savefig()
fig.savefig('my_plot.png')
```

```python
# (Straight line curve)
# Import NumPy as np
import numpy as np
# Create an array of 100 evenly spaced numbers between 0 and 100 using NumPy and
# save it to variable X
X = np.linspace(0,100,100)
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)
fig, ax = plt.subplots()
ax.plot(X,X**2)
```

```python
# (Scatter plots)
# Create a scatter plot of X versus the exponential of X (np.exp(X))
fig, ax = plt.subplots()
ax.scatter(X,np.exp(X))
```

```python
# Create a scatter plot of X versus np.sin(X)
fig, ax = plt.subplots()
ax.scatter(X,np.sin(X))
```

```python
# (Bar plot)
# Create a Python dictionary of 3 of your favourite foods with
# The keys of the dictionary should be the food name and the values their price
favourite_foods = {"Rice": 20.00, "Beans": 10.00, "Corn": 30.00}
print(favourite_foods)
```

```python
# Create a bar graph where the x-axis is the keys of the dictionary
# and the y-axis is the values of the dictionary
fig, ax = plt.subplots()
ax.bar(favourite_foods.keys(), favourite_foods.values())

# Add a title, xlabel and ylabel to the plot
ax.set(title='Favourite_foods', xlabel='Foods', ylabel='Price Value')
```

```python
# Make the same plot as above, except this time make the bars go horizontal
fig, ax = plt.subplots()
ax.barh(favourite_foods.keys(), favourite_foods.values())

ax.set(title='Favourite_foods', xlabel='Price Value', ylabel='Foods')
```

```python
# (Histogram Plots)
# Create a random NumPy array of 1000 normally distributed numbers using NumPy
# and save it to X
X = np.random.normal(size=1000)

# Create a histogram plot of X
fig, ax = plt.subplots()
ax.hist(X)
```

```python
# (Subplots - A figure with multiple plots on it)
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
fig, ax = plt.subplots(2,2)
```

```python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
fig,ax =plt.subplots(2,2, figsize=(10,5))

# Plot X versus X/2 on the top left axes
ax[0,0].plot(X, X/2)

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
ax[0,1].scatter(np.random.rand(10), np.random.rand(10))

# Plot a bar graph of the favourite food keys and values on the bottom left
# subplot
ax[1,0].bar(favourite_foods.keys(), favourite_foods.values())

# Plot a histogram of 1000 random normally distributed numbers on the bottom
# right subplot
ax[1,1].hist(np.random.normal(size=1000))
```

```python
# Try to plot the 'Price' column using the plot() function - it did not work

# Remove the symbols, the final two numbers from the 'Price' column and convert
# it to numbers

# car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '').astype(int)

car_sales['Price'].plot()
```

```python
# Alternatively

# Step 1: First convert the Price column to numeric by removing $ and comma,
# then convert to float.To prevent TypeError: no numeric data to plot.


car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)


# Step 2: Plot the Price column
car_sales.Price.plot()
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the
# 'Price' column
car_sales['Total sales'] = car_sales['Price'].cumsum()
car_sales

# Add a column called 'Sale Date' which lists a series of successive dates
# starting from today (your today)
car_sales['Sale Date'] = pd.date_range('2025-07-14', periods=len(car_sales))

# View the car_sales DataFrame
car_sales
```

```python
# (Visualize Total Sale and Sale Date numeric Columns)

# Use the plot() function to plot the 'Sale Date' column versus
# the 'Total sales' column
car_sales.plot(x='Sale Date', y='Total sales')
```

```python
# Convert the 'Price' column to the integers
car_sales['Price'] = car_sales['Price'].astype(int)

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the
# plot() function
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter')
```

```python
# Import "../data/heart-disease.csv" and save it to the variable "heart_disease"
heart_disease = pd.read_csv('/content/drive/My Drive/data/heart-disease.csv')
heart_disease.head()
```

```python
# View the first 10 rows of the heart_disease DataFrame
heart_disease.head(10)
```

```python
# Create a histogram of the "age" column with 50 bins
heart_disease['age'].plot(kind='hist', bins=50)
```

```python
# Call plot.hist() on the heart_disease DataFrame and toggle the
# "subplots" parameter to True
heart_disease.plot.hist(subplots=True)
```

```python
# Plot looks squished, change figsize
# Call the same line of code from above except change the "figsize" parameter
# to be (10, 30)
heart_disease.plot.hist(subplots=True, figsize=(10,30))
```

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
patients_over_50 = heart_disease[heart_disease['age']>50]

# Create the plot
patients_over_50.plot.hist(figsize=(10,30), subplots=True)

# Plot the data
plt.show()

# Customize the plot
ax = patients_over_50.plot.hist(figsize=(10, 30), subplots=True, bins=15,
    color='skyblue', edgecolor='black', grid=True)

plt.suptitle('Distributions of Features for Patients Over 50', fontsize=16)  # overall title
plt.ylabel('Frequency')

plt.show()

# Add a meanline
patients_over_50.plot.hist(figsize=(10,30), subplots=True)
plt.axvline(patients_over_50['age'].mean(), color='pink')
```
