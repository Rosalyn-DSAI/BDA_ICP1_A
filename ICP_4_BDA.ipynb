```python
# 7. Show the basic statistical description about the data
statistical_description = df.describe()
print(statistical_description)
```

```python
# 8. Check if the data has null values.
df.isnull()
```

```python
# 8a. Replace the null values with the mean
# Compute the percolumn means
means = df.mean()

# Fill nulls with the percolumn means
df.fillna(means, inplace=True)
```

```python
# 9. Select at least two columns and aggregate the data using: min, max, count, mean
# 1. Select 2 columns
subset = df[['Duration', 'Pulse']]

# 2. Apply min, max, count, mean
aggregate_stats = subset.agg(['min', 'max', 'count', 'mean'])

print(aggregate_stats)
```

```python
# 10. Filter the dataframe to select the rows with calories values
# between 500 and 1000
filtered = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print(filtered)
```

```python
# 11. Filter the dataframe to select the rows with calories values > 500
# and pulse < 100.
filtered = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print(filtered)
```

```python
# 12. Create a new df_modified dataframe that contains all the columns
# from df except for Maxpulse.

# we are dropping Maxpulse from dataframe called df.
df_modified = df.drop(columns=['Maxpulse'])

print(df_modified.head())
```

```python
# 13. Delete the Maxpulse column from the main df dataframe
df.drop(columns=['Maxpulse'], inplace=True)
print(df.columns)
```

```python
# 14. Convert the datatype of Calories column to int datatype.

# -> Using astype which is used since there are no null values
df['Calories'] = df['Calories'].astype(int)

print(df['Calories'].dtype)
print(df.head())
```

```python
# 15. Using pandas create a scatter plot for the two columns
# (Duration and Calories).
import matplotlib.pyplot as plt
ax = df.plot.scatter(x='Duration', y='Calories', title='Duration vs Calories')
ax.set_xlabel('Duration')
ax.set_ylabel('Calories')
plt.show()
```
